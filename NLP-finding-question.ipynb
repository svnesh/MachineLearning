```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import nltk
import re
import warnings
warnings.filterwarnings('ignore')
pd.options.display.max_rows = 75
```

```python
df_full = pd.read_csv('sample.csv')
```

```python
df = df_full[['text']]
```

```python
def changeToLower(text):
    return str(text).lower()

df['text_lower'] = df['text'].apply(lambda text:changeToLower(text))
df.head()
```

```python
def clearnonascii(text):
    return ''.join([i if ord(i) < 128 else '' for i in text])

df['text_withoutasci'] = df['text_lower'].apply(lambda text:clearnonascii(text))
df.head()
```

```python
df.drop(columns=['text_lower'], inplace=True)
df.head()
```

```python
twitacc = re.compile('(@\w+\s)')
httpsite = re.compile('((http|www.)\S+)')
def removeTwitacc(text):
    str1 = twitacc.sub('', str(text))
    return httpsite.sub('URL', str1)

df['text_withoutlinkid'] = df['text_withoutasci'].apply(lambda text:removeTwitacc(text))
df.head()
```

```python
import string
PUNCTUATIONS = string.punctuation

def removepunct(text):
    return text.translate(text.maketrans('','', PUNCTUATIONS))

df['text_withoutpunct'] = df['text_withoutlinkid'].apply(lambda text:removepunct(text))
df.head()
```
